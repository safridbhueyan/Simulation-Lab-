# Necessary Libraries

```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

```python
# Given data
data = [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]

# Hypothesized mean
mu = 5.5

print("Data:", data)
print("Hypothesized Mean:", mu)
```

# Calculate Simple Statistic

```python
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", n)
```

# Perform One-Sample t-Test

```python
t_statistic, p_value = stats.ttest_1samp(data, mu)

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

# Decision at α = 0.05

```python
alpha = 0.05

print("Interpretation at 0.05 Significance Level:")

if p_value < alpha:
    print("Decision: Reject the null hypothesis (H0)")
    print("Conclusion: The mean is significantly different from 5.5")
else:
    print("Decision: Fail to reject the null hypothesis (H0)")
    print("Conclusion: No significant difference from 5.5")
```

# Visualization

```python
# Plot distribution
sns.histplot(data, kde=True)

# Mark sample mean
plt.axvline(sample_mean, color='blue', linestyle='--', label='Sample Mean')

# Mark hypothesized mean
plt.axvline(mu, color='red', linestyle='--', label='Hypothesized Mean')

plt.title("Data Distribution with Means")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

# t-distribution
df = n - 1
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x, df)

plt.plot(x, y)
plt.axvline(t_statistic, linestyle='--')

plt.title("t-Distribution with Test Statistic")
plt.xlabel("t value")
plt.ylabel("Density")

plt.show()
```

- Sample mean ≈ 5.4
- Hypothesized mean = 5.5  
- t-statistic ≈ -1.41  
- p-value ≈ 0.22  

Since p-value (0.22) > 0.05, we fail to reject the null hypothesis.

Interpretation:  
There is not enough statistical evidence to conclude that the true mean is significantly different from 5.5.
